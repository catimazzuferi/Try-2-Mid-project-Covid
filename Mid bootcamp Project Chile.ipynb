# Scale Numericals

```python
from sklearn.preprocessing import MinMaxScaler 
from sklearn.preprocessing import StandardScaler


X2 = numerical2

```

```python
transformer = MinMaxScaler().fit(X2)
x2_normalized = transformer.transform(X2)
print(x2_normalized.shape)
x2_normalized = pd.DataFrame(x2_normalized, columns=X2.columns)
```

```python
plt.plot(bigdata2['date'], x2_normalized['people_fully_vaccinated'], label='people_fully_vaccinated')
plt.plot(bigdata2['date'], x2_normalized['new_deaths'], label='new_deaths')

plt.xlabel('date')
plt.ylabel('Valor')
plt.title('Gráfico de nuevas muerte y gente full vacunada')

plt.show()
```

```python
plt.plot(bigdata2['date'], x2_normalized['n_patiens_ICU_that_day'], label='n_patiens_ICU_that_day')
plt.plot(bigdata2['date'], x2_normalized['new_deaths'], label='new_deaths')

plt.xlabel('date')
plt.ylabel('Valor')
plt.title('Gráfico de nuevas muerte y gente en la UCI ese dia')

plt.show()
```

```python
plt.plot(bigdata2['date'], x2_normalized['n_patiens_ICU_that_day'], label='n_patiens_ICU_that_day', marker='o')
plt.plot(bigdata2['date'], x2_normalized['people_fully_vaccinated'], label='people_fully_vaccinated', marker='o')

plt.xlabel('date')
plt.ylabel('Valor')
plt.title('Gráfico de gente full vacunada y gente en la UCI ese dia')

plt.show()
```
